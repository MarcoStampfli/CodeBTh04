```python
import os
import re

# Ordner mit den Dateien (aktuell)
ordner_pfad = "."
praefix = "Test_"

# Alle passenden Dateien sammeln
dateien = sorted([
    f for f in os.listdir(ordner_pfad)
    if f.startswith(praefix) and os.path.isfile(os.path.join(ordner_pfad, f))
])

# Umbenennen mit fortlaufender ID
for index, dateiname in enumerate(dateien, start=1):
    # "Test_" entfernen
    name_ohne_praefix = dateiname[len(praefix):]
    
    # Trenne Dateiname und Endung
    name, ext = os.path.splitext(name_ohne_praefix)
    
    # Suffix wie "_1", "_2" usw. am Ende entfernen
    name = re.sub(r"_\d+$", "", name)
    
    # ID vorne anhängen
    neuer_name = f"{index}_{name}{ext}"
    
    # Pfade erstellen
    alter_pfad = os.path.join(ordner_pfad, dateiname)
    neuer_pfad = os.path.join(ordner_pfad, neuer_name)
    
    # Umbenennen
    os.rename(alter_pfad, neuer_pfad)
    print(f"Umbenannt: {dateiname} -> {neuer_name}")
```

```python
import os
import shutil
import re

# Pfad zum Ordner mit den Dateien
ordner_pfad = "."
ziel_basis = "sortiert"  # Optionaler Hauptordner

# Zielbasis erstellen (optional)
os.makedirs(ziel_basis, exist_ok=True)

# Alle .las-Dateien durchgehen
for dateiname in os.listdir(ordner_pfad):
    if dateiname.endswith(".las") and os.path.isfile(os.path.join(ordner_pfad, dateiname)):
        
        # Versuche den Dachtyp aus dem Dateinamen zu extrahieren
        match = re.search(r"_(\w+)\.las$", dateiname)
        if match:
            dachtyp = match.group(1)  # z.B. "Walmdach", "Flachdach", etc.
            zielordner = os.path.join(ziel_basis, dachtyp)
            os.makedirs(zielordner, exist_ok=True)  # Ordner anlegen, falls nötig
            
            # Datei verschieben
            quelle = os.path.join(ordner_pfad, dateiname)
            ziel = os.path.join(zielordner, dateiname)
            shutil.move(quelle, ziel)
            print(f"Verschoben: {dateiname} -> {zielordner}")
```

```python
import os
import shutil
import re

# Aktueller Pfad
basis_pfad = "./sortiert"

# Alle Einträge im Ordner durchgehen
for ordner in os.listdir(basis_pfad):
    voller_pfad = os.path.join(basis_pfad, ordner)
    
    # Nur Ordner bearbeiten
    if os.path.isdir(voller_pfad):
        # Versuche den reinen Dachtyp am Ende zu extrahieren
        match = re.search(r"([A-Za-z]+dach)$", ordner, re.IGNORECASE)
        if match:
            dachtyp = match.group(1).capitalize()  # z.B. "Satteldach", "Walmdach"
            
            zielordner = os.path.join(basis_pfad, dachtyp)
            os.makedirs(zielordner, exist_ok=True)

            # Zielpfad inkl. Unterordnername (z. B. Walmdach/2_A_Walmdach)
            ziel_unterordner = os.path.join(zielordner, ordner)
            shutil.move(voller_pfad, ziel_unterordner)
            print(f"Verschoben: {ordner} -> {zielordner}")
```

```python
import os
import pandas as pd

# Pfad zum Hauptverzeichnis anpassen
hauptverzeichnis = r".\Satteldach"


# Excel-Datei zur Ausgabe
ausgabedatei = os.path.join(hauptverzeichnis, "Dateilisten.xlsx")

# ExcelWriter für mehrere Tabellenblätter
with pd.ExcelWriter(ausgabedatei, engine='xlsxwriter') as writer:
    for ordnername in os.listdir(hauptverzeichnis):
        ordnerpfad = os.path.join(hauptverzeichnis, ordnername)
        if os.path.isdir(ordnerpfad):
            dateinamen = [
                os.path.splitext(datei)[0]
                for datei in os.listdir(ordnerpfad)
                if os.path.isfile(os.path.join(ordnerpfad, datei))
            ]
            df = pd.DataFrame(dateinamen, columns=["Dateiname_ohne_Endung"])
            # Tabellenblattname muss max. 31 Zeichen haben
            tabellenblattname = ordnername[:31]
            df.to_excel(writer, sheet_name=tabellenblattname, index=False)

print("Excel-Datei erfolgreich erstellt:", ausgabedatei)
```
